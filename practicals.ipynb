```python
import pandas as pd
import nltk
from nltk.parse.corenlp import transform
```

# tokenization

```python
d = pd.read_csv(r"C:\Users\HP\PycharmProjectsAI\NLP\Twitter_Data.csv.xls")
df = d['clean_text'][199]
df
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize

words = word_tokenize(df)
words
```

```python
sent = sent_tokenize(df)
sent
```

# stop word removal

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

sw = set(stopwords.words('english'))

nw = [word for word in words if word not in sw]
nw
```

# stemming & lemmatization

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer

ps = PorterStemmer()

for word in words:
    print(word, "---------->", ps.stem(word))
```

```python
lm = WordNetLemmatizer()

words = ['travesty',
 'modi',
 '’',
 'crook',
 'facts',
 'can',
 '’',
 'suppressedreality',
 'comes',
 'out',
 'speak',
 'for',
 'itself']

for word in words:
    print(word, "----------->", lm.lemmatize(word))
```

# vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

co = CountVectorizer()

x = co.fit_transform(sent).toarray()

print(co.get_feature_names_out())
print(x)
```

```python
tf = TfidfVectorizer()
y = tf.fit_transform(sent).toarray()
print(tf.get_feature_names_out())
print(y)
```
