```python
import nltk
nltk.download('wordnet')
import spacy

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
lm = WordNetLemmatizer()

text = "Iron Man, also known as Tony Stark, is one of the central heroes in the Avengers, blending genius-level engineering with bold, often sarcastic leadership. Using his advanced armored suits, he protects Earth from powerful threats while constantly evolving his technology to face new challenges. Beyond the action, his journey in the Avengers highlights personal growth — from a self-focused inventor to a hero willing to sacrifice everything for the greater good, making him one of the most compelling figures in the team."

words = word_tokenize(text)

for word in words:
    print(word, "==", lm.lemmatize(word))
```

### stemming and lemmatization

```python
text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability. Rising to fame at FC Barcelona, he broke numerous records and won multiple Ballon d’Or awards, becoming a symbol of consistency and creativity on the field. His career reached a historic peak when he led Argentina to victory in the 2022 FIFA World Cup, showcasing leadership and resilience while inspiring millions of fans around the world."

words = word_tokenize(text)
print(words)
```

```python
from nltk.stem import PorterStemmer
ps= PorterStemmer()

for word in words:
    print(word, "==", ps.stem(word))
```

## after stemming

```python
for word in words:
    print(word,"==", lm.lemmatize(word))
```

# lemmatization with pos

```python
# pos = n,v,a,r

for word in words:
    print(word,"==", lm.lemmatize(word,pos='n'))
```

```python
for word in words:
    print(word,"==", lm.lemmatize(word,pos='v'))
```

```python
for word in words:
    print(word,"==", lm.lemmatize(word,pos='a'))
```

```python
for word in words:
    print(word,"==", lm.lemmatize(word,pos='r'))
```

### custamization in nlp using pipe names

```python
nlp = spacy.load("en_core_web_sm")
print(nlp.pipe_names)
```

```python
text = "my broh is shinas , sometimes a call him bro also"

ar = nlp.get_pipe('attribute_ruler')
ar.add([[{"TEXT":"bro"}],[{"TEXT":"broh"}]], {"LEMMA":"BROTHER"})

doc = nlp(text)

for token in doc:
    print(token.text, "==", token.lemma_ )
```

## lemmatization with spcay

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text= "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability. Rising to fame at FC Barcelona, he broke numerous records and won multiple Ballon d’Or awards, becoming a symbol of consistency and creativity on the field. His career reached a historic peak when he led Argentina to victory in the 2022 FIFA World Cup, showcasing leadership and resilience while inspiring millions of fans around the world."

doc = nlp(text)

for token in doc:
    print(token.text, "==", token.lemma_)
```
