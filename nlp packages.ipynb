```python

import nltk

from nltk.tokenize import word_tokenize

text = "nlp is amazing and powerful programming language."
tokens=word_tokenize(text)

print(tokens)
```

```python
from nltk.tokenize import sent_tokenize

text= "nlp is amazing and powerful programming language. it is amazing"

tokens=sent_tokenize(text)

print(tokens)
```

```python
import spacy


nlp=spacy.load("en_core_web_sm")
doc= nlp("nlp is amazing and powerful.")

for token in doc:
    print(token.text)
```

```python
doc = nlp("nlp is amazing. is powerful")

for sentence in doc.sents:
    print(sentence)
```

```python
doc = nlp("my name is shaheer. i am 20 years old . iam son of mr.basheer")

for sentence in doc.sents:
    print(sentence)
```
