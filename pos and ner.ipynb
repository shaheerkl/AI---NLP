# POS with nltk

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability."

words = word_tokenize(text)
print(words)
```

```python
pos_tags = nltk.pos_tag(words)

print(pos_tags)
```

# POS with spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability."

doc = nlp(text)
```

```python
for token in doc:
    print(token.text, "==", token.pos_)
```

# NER with nltk

```python
import nltk
nltk.download('punkt')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
```

```python
text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability."

words = word_tokenize(text)
```

```python
pos_tags = nltk.pos_tag(words)
```

```python
ner = nltk.ne_chunk(pos_tags)
print(ner)
```

# NER with spacy

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
text = 'Lionel Messi plays for Inter Miami in the United States.Lionel Messi won the FIFA World Cup with Argentina in 2022.'

doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, "==", ent.label_)
```
