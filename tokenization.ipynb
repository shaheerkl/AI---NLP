### using nltk

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

```

```python
text = "my name is shaheer, i am coming from kannur. i am 20 years old"

# sentence tokenization
sentences= sent_tokenize(text)
print(sentences)
```

```python
# word tokenization

words=word_tokenize(text)
print(words)
```

### using spacy

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "my name is shaheer, i am coming from kannur. i am 20 years old"
doc=nlp(text)
```

```python
# sentence tokenization
for sent in doc.sents:
    print(sent.text)
```

```python
# word tokenization

for token in doc:
    print(token.text)
```

```python
# character tokenization

text = "shaheer"
chars = list(text)
print(chars)
```
