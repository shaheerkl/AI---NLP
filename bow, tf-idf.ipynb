```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
import pandas as pd
```

```python
docs = [
    "I love machine learning",
    "Machine learning is powerful",
    "I love coding"
]
```

```python
co = CountVectorizer()
x = co.fit_transform(docs)
x
```

```python
tf = TfidfVectorizer()
x = tf.fit_transform(docs).toarray()
x
```
