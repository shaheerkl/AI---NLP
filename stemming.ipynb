```python
import nltk
import spacy
```

### porter stemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']

for word in words:
    print(word,"==", ps.stem(word))
```

### regex stemmer

```python
from nltk.stem import RegexpStemmer
rs= RegexpStemmer('ing$|s$|ed$|ed$|ent$', min=4)


words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']

for word in words:
    print(word, "==", rs.stem(word))
```

### snowball stemmer

```python
from nltk.stem import SnowballStemmer
sn = SnowballStemmer('english')

words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']


for word in words:
    print(word, "==", sn.stem(word))
```
