### porter stemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']

for word in words:
    print(word,"==", ps.stem(word))
```

### regex stemmer

```python
from nltk.stem import RegexpStemmer
rs= RegexpStemmer('ing$|s$|ed$|ed$|ent$', min=4)


words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']

for word in words:
    print(word, "==", rs.stem(word))
```

### snowball stemmer

```python
from nltk.stem import SnowballStemmer
sn = SnowballStemmer('english')

words= ['playing','running','swimming', 'reading','played','studied','walked','classes','boxes','wishes','sadness','development','information']


for word in words:
    print(word, "==", sn.stem(word))
```

### tokenization + stemming

```python
import nltk
import spacy

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer,SnowballStemmer,RegexpStemmer


```

```python
text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability. Rising to fame at FC Barcelona, he broke numerous records and won multiple Ballon d’Or awards, becoming a symbol of consistency and creativity on the field. His career reached a historic peak when he led Argentina to victory in the 2022 FIFA World Cup, showcasing leadership and resilience while inspiring millions of fans around the world."


words = word_tokenize(text)
words
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp(text)

for token in doc:
    print(token.text)
```

```python
ps = PorterStemmer()

for word in words:
    print(word, "==", ps.stem(word))
```

```python
sn = SnowballStemmer('english')

for word in words:
    print(word, "==", sn.stem(word))
```

```python
rs = RegexpStemmer("ing$|ly$|est$|ous$", min=4)

for word in words:
    print(word, "==", rs.stem(word))
```
