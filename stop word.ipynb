```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')

text = "Lionel Messi is widely regarded as one of the greatest footballers in history, known for his extraordinary dribbling, vision, and goal-scoring ability. Rising to fame at FC Barcelona, he broke numerous records and won multiple Ballon d’Or awards, becoming a symbol of consistency and creativity on the field. His career reached a historic peak when he led Argentina to victory in the 2022 FIFA World Cup, showcasing leadership and resilience while inspiring millions of fans around the world."
```

```python
words = word_tokenize(text)

# loading stop words
stop_words = set(stopwords.words('english'))
```

```python
# removing stopwords

filtered_words = [word for word in words if word not in stop_words]
print(filtered_words)
```

```python
stop_words = [token.text for token in doc if not token.is_stop]
print(stop_words)
```

# stop word removal on datasets

```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv(r"C:\Users\HP\PycharmProjects\JupyterProject\Downloads\spam.csv",encoding="latin-1")

df.head()
```

```python
def remove_stopwords(text):
    doc = nlp(text)
    filtered_tokens = [
        token.text
        for token in doc
        if not token.is_stop and not token.is_punct
    ]
    return " ".join(filtered_tokens)
```

```python
df["v2_new"] = df["v2"].apply(remove_stopwords)
```
